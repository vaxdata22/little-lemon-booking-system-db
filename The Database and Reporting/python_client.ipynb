```python
#full name and contact details for every customer that has placed an order greater than $60
query = """SELECT c.CustomerName,c.CustomerContact
            FROM Customers AS c
            INNER JOIN Bookings AS b ON b.CustomerID = c.CustomerID
            INNER JOIN Orders AS o ON b.BookingID = o.BookingID
            WHERE CAST(REPLACE(o.TotalCost,'$','') AS DECIMAL) > 60"""
cursor.execute(query)
results = cursor.fetchall()
print(cursor.column_names)
for _ in results:
    print(_)
```
